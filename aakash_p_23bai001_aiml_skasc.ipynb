```python
#Radial Plot
import numpy as np
categories = df.columns[:-1]  # Exclude species column
averages = df.groupby('species').mean()

for species, avg in averages.iterrows():
    values = avg.values
    values = np.append(values, values[0])  # Close the circle
    angles = np.linspace(0, 2 * np.pi, len(values), endpoint=True)

    plt.polar(angles, values, label=species)
plt.title("Radial Plot of Feature Averages by Species")
plt.legend()
plt.show()
```

```python
#3D plot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
species = {'setosa': 'r', 'versicolor': 'g', 'virginica': 'b'}
for sp in species:
    subset = df[df['species'] == sp]
    ax.scatter(subset['sepal_length'], subset['sepal_width'], subset['petal_length'],
               c=species[sp], label=sp)
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
plt.title("3D Scatter Plot of Iris Features")
plt.legend()
plt.show()
```
